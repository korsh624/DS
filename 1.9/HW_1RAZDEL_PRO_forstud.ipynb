# Итоговый проект.

Анализ личности клиента — это подробный анализ клиентов компании. Он помогает компании лучше понять своих клиентов и облегчает изменение продукции в соответствии со специфическими потребностями, поведением различных типов клиентов.

Анализ личности клиента помогает компании модифицировать свой продукт в соответствии с целевыми потребителями из различных сегментов потребителей. Например, вместо того, чтобы тратить деньги на маркетинг нового продукта для каждого клиента в базе данных компании, компания может проанализировать, какой сегмент клиентов с наибольшей вероятностью купит продукт, и затем продвигать продукт только для этого сегмента.

Описание столбцов:

|Название| Описание |
|--|--|

|MntFruits| Сумма, потраченная на фрукты за последние 2 года|

|AcceptedCmp1| 1, если клиент принял предложение в 1-й кампании, 0 — иначе|
|AcceptedCmp2| 1, если клиент принял предложение во 2-й компании, 0 — иначе|
|AcceptedCmp3| 1, если клиент принял предложение в 3-й кампании, 0 — иначе|
|AcceptedCmp4| 1, если клиент принял предложение в 4-й кампании, 0 — иначе|
|AcceptedCmp5| 1, если клиент принял предложение в 5-й кампании, 0 — иначе|
|Response| 1, если клиент принял предложение в последней кампании, 0 — иначе|
|NumWebPurchases| Количество покупок, совершенных через веб-сайт компании|
|NumCatalogPurchases| Количество покупок, сделанных с помощью каталога|
|NumStorePurchases| Количество покупок, сделанных непосредственно в магазинах|
|NumWebVisitsMonth| Количество посещений веб-сайта компании за последний месяц|

2. Какой средний возраст покупателя? Зависит ли возраст клиента от его покупок во всех категориях?


```python
average_age = 2024 - df['Year_Birth'].mean()

print("Средний возраст покупателя: {:.2f} лет.".format(average_age))
```

```python
age_purchase_sum = df.groupby('Year_Birth')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 
                                            'MntSweetProducts', 'MntGoldProds']].sum()

# Построение графика
plt.figure(figsize=(12, 6))
age_purchase_sum.plot(kind='line', marker='o')
plt.title('Зависимость возраста клиента от покупок в различных категориях')
plt.xlabel('Год рождения')
plt.ylabel('Сумма покупок')
plt.grid(True)
plt.legend(title='Категории покупок')
plt.tight_layout()
plt.show()
```

Средний возраст 55 лет. Покупатели среднего возраста тратят больше всего практически во всех категориях

3. Тратят ли клиент больше денег, если у него много детей и иждивенцев?


```python
df['Total_Purchase']=df[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
df['Numchild']=df[['Kidhome','Teenhome']].sum(axis=1)
df
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Numchild'], df['Total_Purchase'], color='skyblue')
plt.xlabel('Количество иждевенцев')
plt.ylabel('Сумма покупок')
plt.title('Гистограмма суммы покупок по количеству детей')
plt.xticks(df['Numchild'])
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Люди без детей тратят больше

4. Зависят ли покупки от образования? Что предпочитают люди чаще покупать в зависимости от образования?

```python
grouped_data = df.groupby('Education')['Total_Purchase'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['Education'], grouped_data['Total_Purchase'])
plt.xlabel('Образование')
plt.ylabel('Общее количество покупок')
plt.title('Зависимость покупок от образования')
plt.xticks(rotation=45)
plt.show()
```

```python
age_purchase_sum = df.groupby('Education')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 
                                            'MntSweetProducts', 'MntGoldProds']].sum()
plt.figure(figsize=(12, 6))
age_purchase_sum.plot(kind='line', marker='o')
plt.title('Зависимость образования клиента от покупок в различных категориях')
plt.xlabel('Образование')
plt.ylabel('Сумма покупок')
plt.grid(True)
plt.legend(title='Категории покупок')
plt.tight_layout()
plt.show()
```

Более всего тратят клиенты из категории Graduation

5. Женатые люди тратят больше или меньше денег на покупку золота?


```python
grouped_data = df.groupby('Marital_Status').sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['Marital_Status'], grouped_data['MntGoldProds'], color='gold')
plt.xlabel('Семейное положение')
plt.ylabel('Покупки золота')
plt.title('Зависимость покупок золота от семейного положения')
plt.grid(True)
plt.show()
```

Люди состоящие в браке больше тратят денег на золото 

6. Люди, которые покупают товар на сайте чаще берут что-то со скидкой?

```python
plt.figure(figsize=(8, 6))
df['NumWebPurchases'].value_counts().plot(kind='bar', color='lime')
df['NumDealsPurchases'].value_counts().plot(kind='bar', color='magenta')
plt.title('Покупки Magenta - со скидкой, lime - через сайт')
plt.xlabel('Товары')
plt.ylabel('Количество людей')
plt.show()
```

Болшинство покупок со скидкой сделаны не через сайт

7. Зависит ли жалоба клиента от его трат?


```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Total_Purchase'], df['Complain'], alpha=0.5)
plt.title('Зависимость трат клиента и наличия жалоб')
plt.xlabel('Траты клиента за последние 2 года')
plt.ylabel('Количество жалоб')
plt.grid(True)
plt.show()
```

Большенство жалоб поступило от людей которые потратили среднее количество денег или меньше среднего. Зависимость обратная
